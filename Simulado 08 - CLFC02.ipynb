**Pergunta 1**

Um desenvolvedor deseja implantar um aplicativo rapidamente na AWS sem criar manualmente os recursos necessários. Qual serviço da AWS atenderá a esses requisitos?

a) Amazon EC2  
b) AWS CodeBuild  
c) Amazon Personalize  
d) **AWS Elastic Beanstalk**  

**Explicação:**  
**Correta:**  
- **AWS Elastic Beanstalk**:
  - O AWS Elastic Beanstalk é um serviço fácil de usar para implantar e dimensionar rapidamente aplicativos e serviços desenvolvidos com Java, .NET, PHP, Node.js, Python, Ruby, Go e Docker em servidores familiares como Apache, Nginx, Passenger e IIS. O desenvolvedor apenas precisa fazer upload do código do aplicativo, e o Elastic Beanstalk cuida automaticamente da implementação dos recursos de infraestrutura necessários (como instâncias do Amazon EC2, balanceadores de carga, escalonamento automático, etc.).

**Incorretas:**  
- Amazon EC2:
  - Embora o Amazon EC2 permita o provisionamento de máquinas virtuais (instâncias), o desenvolvedor ainda teria que configurar manualmente a infraestrutura, incluindo a criação e configuração de instâncias, balanceadores de carga, grupos de escalonamento automático, etc.
- AWS CodeBuild:
  - O AWS CodeBuild é um serviço de integração contínua que compila o código-fonte, executa testes e produz pacotes de software prontos para implantação. Ele é mais voltado para o processo de construção do aplicativo do que para a implantação automática de infraestrutura.
- Amazon Personalize:
  - O Amazon Personalize é um serviço de machine learning que permite criar recomendações personalizadas para usuários de aplicativos. Ele não é usado para implantar aplicativos.

**Pergunta 2**

Uma empresa está executando um aplicativo local monolítico que não é escalonável e é difícil de manter. A empresa tem um plano para migrar o aplicativo para AWS e dividir o aplicativo em microsserviços. Qual prática recomendada do AWS Well-Architected Framework a empresa está seguindo com este plano?

a) Integrate functional testing as part of AWS deployment (Integrar testes funcionais como parte da implantação na AWS)  
b) Deploy the application to multiple locations (Implantar o aplicativo em várias localizações)  
c) Use automation to deploy changes (Usar automação para implantar mudanças)  
d) **Implement loosely coupled dependencies (Implementar dependências fracamente acopladas)**  

**Explicação:**  
**Correta:**  
- **Implement loosely coupled dependencies (Implementar dependências fracamente acopladas)**:
  - Dividir um aplicativo monolítico em microsserviços segue a prática recomendada de implementar dependências fracamente acopladas. Os microsserviços são componentes independentes que podem ser desenvolvidos, implantados e escalonados de forma independente, o que aumenta a flexibilidade e facilita a manutenção e a escalabilidade do sistema.

**Incorretas:**  
- Integrate functional testing as part of AWS deployment (Integrar testes funcionais como parte da implantação na AWS):
  - Embora seja uma prática recomendada importante, a integração de testes funcionais não está diretamente relacionada ao processo de dividir um aplicativo monolítico em microsserviços.
- Deploy the application to multiple locations (Implantar o aplicativo em várias localizações):
  - Implantar o aplicativo em várias localizações ajuda na disponibilidade e recuperação de desastres, mas não está diretamente relacionado à divisão de um aplicativo monolítico em microsserviços.
- Use automation to deploy changes (Usar automação para implantar mudanças):
  - A automação de implantações é uma prática recomendada para garantir a consistência e reduzir o erro humano, mas não aborda especificamente a migração de um monolito para microsserviços.

**Pergunta 3**

Qual opção é uma localização física da infraestrutura global da AWS?

a) AWS DataSync (Sincronização de Dados da AWS)  
b) AWS Organizations (Organizações da AWS)  
c) Amazon Connect (Amazon Connect)  
d) **AWS Region (Região da AWS)**  

**Explicação:**  
**Correta:**  
- **AWS Region (Região da AWS)**:
  - AWS Region (Região da AWS) é uma localização física em uma área geográfica específica onde a AWS agrupa seus data centers. Cada região é composta por várias zonas de disponibilidade (Availability Zones), que são locais distintos e isolados dentro da região.

**Incorretas:**  
- AWS DataSync:
  - AWS DataSync é um serviço de transferência de dados que automatiza a movimentação de dados entre armazenamento local e serviços de armazenamento da AWS, mas não é uma localização física.
- AWS Organizations:
  - AWS Organizations é um serviço que permite que você gerencie e controle múltiplas contas da AWS centralmente, mas não é uma localização física.
- Amazon Connect:
  - Amazon Connect é um serviço de central de contato baseado na nuvem que permite que as empresas configurem e gerenciem contact centers virtuais, mas também não é uma localização física.

**Pergunta 4**

Um profissional de nuvem precisa obter relatórios de conformidade da AWS antes de migrar um ambiente para a nuvem AWS. Como esses relatórios podem ser gerados?

a) Open a case with AWS Support (Abrir um caso com o Suporte da AWS)  
b) Generate the reports with Amazon Macie (Gerar os relatórios com Amazon Macie)  
c) Contact the AWS Compliance team (Contactar a equipe de Conformidade da AWS)  
d) **Download the reports from AWS Artifact (Baixar os relatórios do AWS Artifact)**  

**Explicação:**  
**Correta:**  
- **Download the reports from AWS Artifact (Baixar os relatórios do AWS Artifact)**:
  - AWS Artifact é um portal de autoatendimento que oferece acesso a uma variedade de relatórios de conformidade, incluindo relatórios de auditoria de terceiros, certificações e outros documentos que demonstram a conformidade da AWS com vários padrões de segurança e conformidade. Esses relatórios podem ser baixados diretamente do AWS Artifact.

**Incorretas:**  
- Open a case with AWS Support:
  - Abrir um caso com o Suporte da AWS não é necessário para obter relatórios de conformidade, pois esses relatórios estão disponíveis para download imediato no AWS Artifact.
- Generate the reports with Amazon Macie:
  - Amazon Macie é um serviço de segurança e privacidade de dados que usa aprendizado de máquina para descobrir, classificar e proteger dados confidenciais, mas não é utilizado para gerar relatórios de conformidade.
- Contact the AWS Compliance team:
  - Embora a equipe de Conformidade da AWS possa fornecer informações adicionais, a maneira mais eficiente e direta de obter os relatórios de conformidade é através do AWS Artifact.

  - A AWS Command Line Interface (CLI) é uma ferramenta que permite interagir com os serviços da AWS a partir da linha de comando, mas não é uma estrutura de programação para definir infraestrutura usando linguagens de programação.

**Pergunta 7**

A elasticidade na Nuvem AWS refere-se a qual das opções a seguir? (Escolha dois.)

a) How quickly an Amazon EC2 instance can be restarted (Quão rapidamente uma instância Amazon EC2 pode ser reiniciada)  
b) The pay-as-you-go billing model (O modelo de cobrança paga conforme o uso)  
c) **The ability to rightsize resources as demand shifts (A capacidade de dimensionar corretamente os recursos conforme a demanda muda)**  
d) The maximum amount of RAM an Amazon EC2 instance can use (A quantidade máxima de RAM que uma instância Amazon EC2 pode usar)  
e) **How easily resources can be procured when they are needed (Quão facilmente os recursos podem ser adquiridos quando necessário)**  

**Explicação:**  
**Corretas:**  
- **The ability to rightsize resources as demand shifts (A capacidade de dimensionar corretamente os recursos conforme a demanda muda)**:
  - A elasticidade refere-se à capacidade de aumentar ou diminuir os recursos de TI conforme necessário para atender à demanda. Isso inclui dimensionar corretamente os recursos para otimizar custos e desempenho conforme a carga de trabalho varia.
- **How easily resources can be procured when they are needed (Quão facilmente os recursos podem ser adquiridos quando necessário)**:
  - A elasticidade também envolve a facilidade com que os recursos podem ser provisionados ou liberados conforme necessário. Isso permite que as empresas respondam rapidamente a mudanças na demanda, garantindo que tenham a capacidade adequada para suas necessidades.

**Incorretas:**  
- How quickly an Amazon EC2 instance can be restarted (Quão rapidamente uma instância Amazon EC2 pode ser reiniciada):
  - A rapidez com que uma instância Amazon EC2 pode ser reiniciada está relacionada à disponibilidade e recuperação, mas não define diretamente a elasticidade.
- The pay-as-you-go billing model (O modelo de cobrança paga conforme o uso):
  - Embora o modelo de cobrança paga conforme o uso seja uma característica importante da computação em nuvem, ele está mais relacionado à flexibilidade financeira do que à elasticidade operacional dos recursos.
- The maximum amount of RAM an Amazon EC2 instance can use (A quantidade máxima de RAM que uma instância Amazon EC2 pode usar):
  - A quantidade máxima de RAM que uma instância EC2 pode usar é uma característica de dimensionamento vertical, não necessariamente relacionada à elasticidade, que geralmente envolve dimensionamento horizontal (adicionando ou removendo instâncias).

**Pergunta 8**

Uma empresa está usando uma plataforma central de dados para gerenciar vários tipos de dados de seus clientes. A empresa deseja usar os serviços da AWS para descobrir, transformar e visualizar os dados. Qual combinação de serviços da AWS a empresa deve usar para atender a esses requisitos? (Escolha dois.)

a) Amazon Quantum Ledger Database (Amazon QLDB)  
b) Amazon Redshift  
c) Amazon Elastic File System (Amazon EFS)  
d) **Amazon QuickSight**  
e) **AWS Glue**  

**Explicação:**  
**Corretas:**  
- **Amazon QuickSight**:
  - Amazon QuickSight é um serviço de business intelligence (BI) que permite criar e publicar dashboards interativos que podem ser acessados de dispositivos móveis e navegadores da web. Ele é ideal para visualizar dados transformados e gerar insights a partir deles.
- **AWS Glue**:
  - AWS Glue é um serviço de ETL (extração, transformação e carregamento) que ajuda a descobrir, preparar e transformar dados para análise. Ele pode catalogar dados de várias fontes e transformá-los para torná-los úteis para análise e visualização.

**Incorretas:**

- Amazon Redshift:
  - Amazon Redshift é um data warehouse escalável que pode ser usado para análises de dados, mas ele não é diretamente utilizado para transformar ou visualizar dados. Ele é mais adequado para armazenar e consultar grandes volumes de dados.
- Amazon Elastic File System (Amazon EFS):
  - Amazon EFS é um serviço de armazenamento de arquivos que permite a criação de sistemas de arquivos elásticos e escaláveis, mas não oferece funcionalidades específicas para descobrir, transformar ou visualizar dados.

**Pergunta 9**

Uma plataforma de e-learning precisa executar um aplicativo durante 2 meses por ano. O aplicativo será implantado em instâncias do Amazon EC2. Qualquer tempo de inatividade do aplicativo durante esses 2 meses deve ser evitado. Qual opção de compra do EC2 atenderá a esses requisitos de maneira MAIS econômica?

a) Reserved Instances (Instâncias Reservadas)  
b) Dedicated Hosts (Hosts Dedicados)  
c) Spot Instances (Instâncias Spot)  
d) **On-Demand Instances (Instâncias Sob Demanda)**  

**Explicação:**  
**Correta:**  
- **On-Demand Instances (Instâncias Sob Demanda)**:
  - Instâncias Sob Demanda permitem que você pague pela capacidade de computação por hora ou por segundo, sem compromisso de longo prazo. Para um uso de 2 meses por ano, esta é a opção mais econômica e flexível, garantindo que o aplicativo esteja sempre disponível durante o período necessário, evitando qualquer tempo de inatividade.

**Incorretas:**  
- Reserved Instances (Instâncias Reservadas):
  - Instâncias Reservadas são mais econômicas para uso contínuo ou previsível ao longo de um ano ou mais, com compromissos de 1 ou 3 anos. Para um uso de apenas 2 meses por ano, isso não seria a opção mais econômica.
- Dedicated Hosts (Hosts Dedicados):
  - Hosts Dedicados são mais caros e são usados principalmente quando você precisa de um servidor físico dedicado para atender a requisitos específicos de conformidade ou licenciamento. Para um uso de 2 meses por ano, essa seria uma opção desnecessariamente cara.
- Spot Instances (Instâncias Spot):
  - Instâncias Spot são oferecidas a preços significativamente mais baixos, mas podem ser interrompidas pela AWS com um aviso de apenas dois minutos se a capacidade for necessária para outra demanda. Isso não atende ao requisito de evitar qualquer tempo de inatividade.

**Pergunta 10**

Qual modelo de preço do Amazon EC2 fornece a MAIOR economia de custos para um servidor de banco de dados sempre ativo e dimensionado corretamente que será executado por um projeto que durará 1 ano?

a) Instâncias reservadas convertíveis  
b) Instâncias spot  
c) Instâncias sob demanda  
d) **Instâncias reservadas padrão**  

**Explicação:**  
**Correta:**  
- **Instâncias reservadas padrão**:
  - Instâncias reservadas padrão oferecem uma economia significativa em relação às instâncias sob demanda ao comprometer-se com um uso consistente por um período de 1 ou 3 anos. Para um projeto que durará 1 ano, as instâncias reservadas padrão são a melhor opção para economizar custos, pois oferecem um desconto significativo em comparação com as instâncias sob demanda.

**Incorretas:**  
- Instâncias reservadas convertíveis:
  - Instâncias reservadas convertíveis permitem trocar a configuração de instância durante o período da reserva, mas geralmente são mais caras do que as instâncias reservadas padrão e não oferecem a maior economia para um uso consistente de 1 ano.
- Instâncias spot:
  - Instâncias spot oferecem preços muito baixos para capacidade de EC2 não utilizada, mas são mais adequadas para cargas de trabalho que podem tolerar interrupções, o que não é ideal para um servidor de banco de dados sempre ativo.
- Instâncias sob demanda:
  - Instâncias sob demanda cobram pelo uso por hora sem compromisso a longo prazo. Elas oferecem flexibilidade, mas não são tão econômicas quanto as instâncias reservadas para um uso constante de 1 ano.

**Pergunta 11**

Uma empresa está usando o Nível Gratuito da AWS para vários serviços AWS em uma aplicação. O que acontecerá se o período de uso do Nível Gratuito expirar ou se o uso da aplicação exceder os limites de uso do Nível Gratuito?

a) A conta AWS da empresa será congelada e poderá ser reiniciada após o estabelecimento de um plano de pagamento.  
b) O Suporte da AWS entrará em contato com a empresa para configurar cobranças de serviço padrão.  
c) A empresa será cobrada pelos serviços consumidos durante o período do Nível Gratuito, além de cobranças adicionais pelo consumo de serviços após o período do Nível Gratuito.  
d) **A empresa será cobrada pelas taxas padrão de serviço pay-as-you-go pelo uso que exceder o uso do Nível Gratuito.**  

**Explicação:**  
**Correta:**  
- **A empresa será cobrada pelas taxas padrão de serviço pay-as-you-go pelo uso que exceder o uso do Nível Gratuito**:
  - Quando o período de uso do Nível Gratuito expirar ou se o uso da aplicação exceder os limites do Nível Gratuito, a empresa será cobrada pelas taxas padrão de serviço pay-as-you-go (pague conforme o uso) da AWS para qualquer uso que exceda o uso do Nível Gratuito. Não há contato prévio do Suporte da AWS ou congelamento da conta devido ao excedente do Nível Gratuito. A cobrança é automática e baseada no uso real dos serviços AWS.

**Incorretas:**  
- A conta AWS da empresa será congelada e poderá ser reiniciada após o estabelecimento de um plano de pagamento:
  - Incorreta. A conta não é congelada se os limites do Nível Gratuito forem excedidos. A cobrança será feita automaticamente com base nas taxas padrão de serviço pay-as-you-go.
- O Suporte da AWS entrará em contato com a empresa para configurar cobranças de serviço padrão:
  - Incorreta. O Suporte da AWS não entra em contato para configurar cobranças. As cobranças são automáticas e baseadas no uso dos serviços.
- A empresa será cobrada pelos serviços consumidos durante o período do Nível Gratuito, além de cobranças adicionais pelo consumo de serviços após o período do Nível Gratuito:
  - Incorreta. Durante o período do Nível Gratuito, a empresa não é cobrada pelos serviços que estão dentro dos limites do Nível Gratuito. Somente o uso que exceder os limites será cobrado.

O Migration Evaluator coleta dados detalhados sobre a infraestrutura existente, o uso e os custos, e fornece análises e relatórios que permitem entender melhor os custos de migração e operação na nuvem. - AWS Cost Explorer:
  - AWS Cost Explorer é uma ferramenta de análise de custos que permite visualizar e analisar os custos de uso da AWS ao longo do tempo. Ele não é projetado para fornecer uma linha de base das operações locais ou estimar os custos de migração.
- AWS Systems Manager Agent (SSM Agent):
  - AWS Systems Manager Agent é um agente instalado em instâncias do Amazon EC2 e servidores locais para gerenciar e automatizar tarefas administrativas. Ele não fornece uma linha de base das operações locais ou estimativas de custos de migração.

**Pergunta 13**

Uma empresa possui um grupo centralizado de usuários com grandes requisitos de armazenamento de arquivos que excederam o espaço disponível no local. A empresa deseja ampliar seus recursos de armazenamento de arquivos para esse grupo, mantendo ao mesmo tempo o benefício de desempenho do compartilhamento local de conteúdo. Qual é a solução AWS MAIS eficiente operacionalmente para este cenário?

a) Crie um bucket do Amazon S3 para cada usuário. Monte cada bucket usando uma utilidade de montagem de sistema de arquivos S3.  
b) Implante uma instância do Amazon EC2 e anexe um volume Amazon Elastic Block Store (Amazon EBS) Provisioned IOPS. Compartilhe o volume EBS diretamente com os usuários.  
c) Mova o ambiente de trabalho de cada usuário para o Amazon WorkSpaces. Configure uma conta do Amazon WorkDocs para cada usuário.  
d) **Configure e implante um AWS Storage Gateway file gateway. Conecte a estação de trabalho de cada usuário ao file gateway.**  

**Explicação:**  
**Correta:**  
- **Configure e implante um AWS Storage Gateway file gateway. Conecte a estação de trabalho de cada usuário ao file gateway**:
  - Esta solução permite a integração do armazenamento local com a nuvem, mantendo o desempenho de acesso aos arquivos mais frequentemente utilizados e oferecendo escalabilidade via Amazon S3.

**Incorretas:**

  - Embora o Amazon EC2 com EBS Provisioned IOPS possa oferecer alto desempenho de armazenamento, compartilhar diretamente um volume EBS entre vários usuários não é prático e pode ser complexo em termos de gerenciamento e segurança.
- Mova o ambiente de trabalho de cada usuário para o Amazon WorkSpaces. Configure uma conta do Amazon WorkDocs para cada usuário:
  - O Amazon WorkSpaces oferece desktops virtuais na nuvem, e o Amazon WorkDocs fornece armazenamento e compartilhamento de arquivos. No entanto, mover todos os ambientes de trabalho para a nuvem pode ser uma solução drástica e pode não atender à necessidade de manter o compartilhamento de arquivos local com alto desempenho.
