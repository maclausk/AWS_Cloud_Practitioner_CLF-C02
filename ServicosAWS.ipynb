## Computação

### Amazon EC2
O Amazon EC2 oferece capacidade computacional escalável na nuvem, permitindo executar diversas cargas de trabalho, incluindo HPC (High Performance Computing).

### AWS Elastic Beanstalk
O AWS Elastic Beanstalk facilita a implantação e o gerenciamento de aplicativos web, gerenciando automaticamente a infraestrutura necessária.

### AWS Lambda
O AWS Lambda permite executar código em resposta a eventos sem a necessidade de gerenciar servidores.

### AWS Batch
O AWS Batch permite executar cargas de trabalho em larga escala na nuvem, ideal para processamento em lote.

### Amazon Lightsail
O Amazon Lightsail é um VPS que facilita a construção e hospedagem de aplicações na nuvem.

### AWS Outposts
O AWS Outposts fornece racks de computação e armazenamento para executar tarefas on-premises, conectando-se à AWS.

## Armazenamento

### Amazon S3
O Amazon S3 oferece armazenamento de objetos escalável e durável, ideal para armazenar e recuperar qualquer quantidade de dados a qualquer momento.

### Amazon EBS
O Amazon EBS fornece armazenamento de blocos persistente para instâncias EC2.

### Amazon EFS
O Amazon EFS oferece um sistema de arquivos elástico para cargas de trabalho em nuvem.

### AWS Storage Gateway
O AWS Storage Gateway integra ambientes on-premises com armazenamento em nuvem, permitindo backup e arquivamento.

### AWS Backup
O AWS Backup permite a criação e o gerenciamento centralizado de backups de serviços AWS.

## Redes e Entrega de Conteúdo

### Amazon VPC
O Amazon VPC permite a criação de redes virtuais isoladas na nuvem AWS.

### AWS Direct Connect
O AWS Direct Connect oferece uma conexão de rede dedicada entre a infraestrutura on-premises e a AWS.

### AWS VPN
O AWS VPN estabelece conexões seguras entre redes locais e a AWS.

## Banco de Dados

### Amazon RDS
O Amazon RDS é um serviço de banco de dados relacional gerenciado que suporta várias opções de mecanismos de banco de dados.

### Amazon DynamoDB
O Amazon DynamoDB é um banco de dados NoSQL totalmente gerenciado, rápido e flexível.

### Amazon Redshift
O Amazon Redshift é um data warehouse rápido e totalmente gerenciado para análise de grandes conjuntos de dados.

### Amazon DocumentDB
O Amazon DocumentDB é um banco de dados NoSQL gerenciado e compatível com MongoDB.

### Amazon Aurora
O Amazon Aurora é um banco de dados relacional compatível com MySQL e PostgreSQL, altamente disponível e escalável.

## Análise

### AWS Glue
O AWS Glue é um serviço de integração de dados sem servidor que facilita a preparação de dados para análise.

### Amazon Athena
O Amazon Athena permite consultas interativas diretamente no Amazon S3 usando SQL padrão.

### Amazon EMR
O Amazon EMR facilita o processamento de grandes volumes de dados usando frameworks como Hadoop e Spark.

### Amazon QuickSight
O Amazon QuickSight permite a criação de dashboards interativos e análises de dados em linguagem natural.

### Amazon Kinesis
O Amazon Kinesis captura, processa e analisa dados em tempo real, facilitando a análise de streaming de dados.

## Machine Learning

### Amazon SageMaker
O Amazon SageMaker permite a criação, treinamento e implantação de modelos de machine learning em escala.

### Amazon Textract
O Amazon Textract usa machine learning para extrair texto e dados de documentos digitalizados.

### Amazon Transcribe
O Amazon Transcribe converte fala em texto usando machine learning.

### Amazon Polly
O Amazon Polly converte texto em fala realista usando machine learning.

### Amazon Translate
O Amazon Translate traduz automaticamente texto entre diferentes idiomas usando machine learning.

## Segurança, Identidade e Conformidade

### AWS IAM
O AWS IAM permite gerenciar o acesso aos serviços e recursos da AWS com segurança.

### AWS KMS
O AWS KMS permite a criação e o gerenciamento de chaves criptográficas usadas para criptografar dados.

### AWS CloudHSM
O AWS CloudHSM oferece armazenamento seguro de chaves criptográficas em hardware.

### AWS Shield
O AWS Shield protege aplicativos contra ataques DDoS.

### AWS WAF
O AWS WAF protege aplicativos web contra explorações comuns da web.

### AWS Macie
O AWS Macie usa machine learning para identificar e proteger dados sensíveis.

### AWS GuardDuty
O AWS GuardDuty detecta atividades maliciosas e anomalias para proteger contas e workloads da AWS.

### AWS Inspector
O AWS Inspector avalia a segurança de aplicativos, procurando vulnerabilidades ou desvios das melhores práticas.

### AWS Secrets Manager
O AWS Secrets Manager gerencia e recupera credenciais de acesso e outros segredos.

### AWS Config
O AWS Config permite avaliar, auditar e avaliar as configurações de recursos da AWS.

## Gerenciamento e Governança

### AWS CloudTrail
O AWS CloudTrail registra todas as chamadas de API para a conta AWS, facilitando auditorias.

### AWS Systems Manager
O AWS Systems Manager oferece um hub de gerenciamento para recursos na AWS e em ambientes híbridos.

### AWS License Manager
O AWS License Manager facilita o gerenciamento de licenças de software.

### AWS Managed Services
O AWS Managed Services oferece recursos proativos, preventivos e de detecção para operações na nuvem.

## Integração de Aplicações

### Amazon SNS
O Amazon SNS é um serviço de mensagem pub/sub que facilita a comunicação entre aplicativos distribuídos.

### Amazon SQS
O Amazon SQS é um serviço de fila de mensagens que desacopla e dimensiona microsserviços.

### Amazon EventBridge
O Amazon EventBridge facilita a criação de arquiteturas orientadas a eventos em escala.

### Amazon API Gateway
O Amazon API Gateway permite a criação, publicação, manutenção, monitoração e proteção de APIs em grande escala.

### AWS Step Functions
O AWS Step Functions orquestra serviços da AWS, automatizando processos de negócios com fluxos de trabalho visuais.

### AWS CodeCommit
O AWS CodeCommit é um serviço de controle de versão que hospeda repositórios Git seguros e escaláveis.

### AWS CodePipeline
O AWS CodePipeline automatiza pipelines de entrega contínua para atualizações rápidas e confiáveis de aplicativos e infraestruturas.

### AWS CodeBuild
O AWS CodeBuild compila código-fonte, executa testes e produz pacotes de software prontos para implantação.

### AWS CodeDeploy
O AWS CodeDeploy automatiza a implantação de código em qualquer instância, incluindo Amazon EC2 e servidores on-premises.

### AWS CodeStar
O AWS CodeStar gerencia o ciclo de vida de desenvolvimento de aplicativos, integrando-se com ferramentas populares de desenvolvimento.

## DevOps

### AWS OpsWorks
O AWS OpsWorks é um serviço de automação de configuração que usa Chef e Puppet para gerenciar infraestruturas em nuvem e on-premises.

### AWS CloudFormation
O AWS CloudFormation permite modelar e configurar seus recursos da AWS para que você possa passar menos tempo gerenciando esses recursos e mais tempo se concentrando nos aplicativos executados neles.

### AWS Service Catalog
O AWS Service Catalog permite que as organizações criem e gerenciem catálogos de serviços de TI aprovados para uso na AWS.
