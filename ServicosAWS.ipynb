## Computação

### Amazon EC2
O Amazon EC2 oferece capacidade computacional escalável na nuuvem, permitindo executar diversas cargas de trabalho, incluindo HPC (High Performance Computing).
- **Exemplo:** Hospedagem de um site de e-commerce que precisa escalar rapidamente durante picos de vendas, como na Black Friday.

### AWS Elastic Beanstalk
O AWS Elastic Beanstalk facilita a implantação e o gerenciamento de aplicativos web, gerenciando automaticamente a infraestrutura necessária.
- **Exemplo:** Desenvolvimento de um blog corporativo onde os desenvolvedores podem focar no código sem se preocupar com a infraestrutura.

### AWS Lambda
O AWS Lambda permite executar código em resposta a eventos sem a necessidade de gerenciar servidores.
- **Exemplo:** Processamento em tempo real de imagens carregadas por usuários em um aplicativo de compartilhamento de fotos.

### AWS Batch
O AWS Batch permite executar cargas de trabalho em larga escala na nuvem, ideal para processamento em lote.
- **Exemplo:** Execução de análises financeiras complexas e geração de relatórios no final do dia.

### Amazon Lightsail
O Amazon Lightsail é um VPS que facilita a construção e hospedagem de aplicações na nuvem.
- **Exemplo:** Hospedagem de um site de portfólio pessoal ou uma aplicação web para pequenas empresas.

### AWS Outposts
O AWS Outposts fornece racks de computação e armazenamento para executar tarefas on-premises, conectando-se à AWS.
- **Exemplo:** Processamento de dados sensíveis localmente em um hospital, mas com integração e backup na nuvem AWS.

### AWS App Runner
O AWS App Runner é um serviço totalmente gerenciado que facilita a criação e a execução de aplicações baseadas em contêineres.
- **Exemplo:** Implantação de uma aplicação web de pequeno porte sem necessidade de gerenciar a infraestrutura subjacente.

### AWS Fargate
O AWS Fargate permite executar contêineres sem a necessidade de gerenciar servidores ou clusters.
- **Exemplo:** Execução de contêineres Docker em um ambiente de microsserviços.

## Armazenamento

### Amazon S3
O Amazon S3 oferece armazenamento de objetos escalável e durável, ideal para armazenar e recuperar qualquer quantidade de dados a qualquer momento.
- **Exemplo:** Armazenamento de vídeos para um serviço de streaming.

### Amazon EBS
O Amazon EBS fornece armazenamento de blocos persistente para instâncias EC2.
- **Exemplo:** Armazenamento de dados de um banco de dados relacional usado por um aplicativo de gestão empresarial.

### Amazon EFS
O Amazon EFS oferece um sistema de arquivos elástico para cargas de trabalho em nuvem.
- **Exemplo:** Compartilhamento de arquivos entre várias instâncias EC2 em um ambiente de desenvolvimento colaborativo.

### AWS Storage Gateway
O AWS Storage Gateway integra ambientes on-premises com armazenamento em nuvem, permitindo backup e arquivamento.
- **Exemplo:** Backup de servidores locais de uma empresa de contabilidade para o Amazon S3.

### AWS Backup
O AWS Backup permite a criação e o gerenciamento centralizado de backups de serviços AWS.
- **Exemplo:** Backup automatizado de bancos de dados RDS e volumes EBS para conformidade regulatória em uma fintech.

### AWS Snowball
O AWS Snowball é um serviço de transferência de dados que permite mover grandes volumes de dados para dentro e fora da AWS usando dispositivos físicos.
- **Exemplo:** Migração de dados de um data center local para a AWS.

### Amazon S3 Glacier
O Amazon S3 Glacier é um serviço de armazenamento de longo prazo para backup e arquivamento.
- **Exemplo:** Armazenamento de dados de conformidade e registros financeiros que precisam ser retidos por longos períodos.

## Redes e Entrega de Conteúdo

### Amazon VPC
O Amazon VPC permite a criação de redes virtuais isoladas na nuvem AWS.
- **Exemplo:** Criação de uma rede privada para hospedar aplicativos seguros de uma empresa de serviços financeiros.

### AWS Direct Connect
O AWS Direct Connect oferece uma conexão de rede dedicada entre a infraestrutura on-premises e a AWS.
- **Exemplo:** Conexão de alta velocidade entre o data center de uma empresa de mídia e a AWS para transferência rápida de arquivos de vídeo.

### AWS VPN
O AWS VPN estabelece conexões seguras entre redes locais e a AWS.
- **Exemplo:** Conexão segura entre a rede corporativa de um escritório remoto e a VPC na AWS.

### Amazon CloudFront
O Amazon CloudFront é um serviço de entrega de conteúdo que distribui dados, vídeos, aplicativos e APIs para clientes globalmente com baixa latência.
- **Exemplo:** Distribuição de um site estático globalmente com rápida entrega de conteúdo.

### AWS Global Accelerator
O AWS Global Accelerator melhora a disponibilidade e a performance dos seus aplicativos globais ao rotear o tráfego dos usuários para os endpoints mais próximos com base na latência.
- **Exemplo:** Melhorar a performance de um serviço de jogos online para jogadores em diferentes regiões.

## Banco de Dados

### Amazon RDS
O Amazon RDS é um serviço de banco de dados relacional gerenciado que suporta várias opções de mecanismos de banco de dados.
- **Exemplo:** Hospedagem de um banco de dados MySQL para um aplicativo de gestão de clientes (CRM).

### Amazon DynamoDB
O Amazon DynamoDB é um banco de dados NoSQL totalmente gerenciado, rápido e flexível.
- **Exemplo:** Armazenamento de sessões de usuário para um jogo online que precisa de baixa latência e alta disponibilidade.

### Amazon Redshift
O Amazon Redshift é um data warehouse rápido e totalmente gerenciado para análise de grandes conjuntos de dados.
- **Exemplo:** Análise de dados de vendas e comportamento do cliente para uma grande rede de varejo.

### Amazon DocumentDB
O Amazon DocumentDB é um banco de dados NoSQL gerenciado e compatível com MongoDB.
- **Exemplo:** Armazenamento de dados de perfis de usuários para um aplicativo de rede social.

### Amazon Aurora
O Amazon Aurora é um banco de dados relacional compatível com MySQL e PostgreSQL, altamente disponível e escalável.
- **Exemplo:** Banco de dados para uma aplicação SaaS de gerenciamento de projetos com milhares de usuários simultâneos.

### Amazon Timestream
O Amazon Timestream é um serviço de banco de dados para séries temporais, ideal para IoT e aplicações operacionais que geram grandes volumes de dados de séries temporais.
- **Exemplo:** Armazenamento e análise de dados de sensores de dispositivos IoT.

### Amazon Neptune
O Amazon Neptune é um serviço de banco de dados gráfico rápido, confiável e totalmente gerenciado.
- **Exemplo:** Gerenciamento de redes sociais e recomendações de produtos baseadas em grafos.

## Análise

### AWS Glue
O AWS Glue é um serviço de integração de dados sem servidor que facilita a preparação de dados para análise.
- **Exemplo:** ETL (extrair, transformar e carregar) de dados de várias fontes para um data warehouse.

### Amazon Athena
O Amazon Athena permite consultas interativas diretamente no Amazon S3 usando SQL padrão.
- **Exemplo:** Consulta de logs de servidor armazenados no S3 para análise de desempenho de um site.

### Amazon EMR
O Amazon EMR facilita o processamento de grandes volumes de dados usando frameworks como Hadoop e Spark.
- **Exemplo:** Processamento de grandes datasets para uma empresa de biotecnologia analisando sequências de DNA.

### Amazon QuickSight
O Amazon QuickSight permite a criação de dashboards interativos e análises de dados em linguagem natural.
- **Exemplo:** Criação de dashboards de vendas e marketing para uma equipe de análise de negócios.

### Amazon Kinesis
O Amazon Kinesis captura, processa e analisa dados em tempo real, facilitando a análise de streaming de dados.
- **Exemplo:** Monitoramento de transações em tempo real para detecção de fraudes em uma plataforma de pagamento online.

## Machine Learning

### Amazon SageMaker
O Amazon SageMaker permite a criação, treinamento e implantação de modelos de machine learning em escala.
- **Exemplo:** Treinamento de modelos de recomendação de produtos para um site de e-commerce.

### Amazon Textract
O Amazon Textract usa machine learning para extrair texto e dados de documentos digitalizados.
- **Exemplo:** Extração automática de dados de formulários médicos digitalizados para um sistema de registros eletrônicos de saúde.

### Amazon Transcribe
O Amazon Transcribe converte fala em texto usando machine learning.
- **Exemplo:** Transcrição automática de chamadas de suporte ao cliente para análise de qualidade.

### Amazon Polly
O Amazon Polly converte texto em fala realista usando machine learning.
- **Exemplo:** Geração de áudio para um assistente virtual em um aplicativo de educação.

### Amazon Translate
O Amazon Translate traduz automaticamente texto entre diferentes idiomas usando machine learning.
- **Exemplo:** Tradução automática de descrições de produtos em um marketplace global.

### Amazon Personalize
O Amazon Personalize oferece recomendações personalizadas utilizando machine learning.
- **Exemplo:** Recomendação de produtos em um site de e-commerce com base no histórico de navegação do usuário.

## Segurança, Identidade e Conformidade

### AWS IAM
O AWS IAM permite gerenciar o acesso aos serviços e recursos da AWS com segurança.
- **Exemplo:** Controle de acesso granular para desenvolvedores e administradores em uma empresa de software.

### AWS KMS
O AWS KMS permite a criação e o gerenciamento de chaves criptográficas usadas para criptografar dados.
- **Exemplo:** Criptografia de dados sensíveis armazenados no Amazon S3 para conformidade com regulamentos de privacidade.

### AWS CloudHSM
O AWS CloudHSM oferece armazenamento seguro de chaves criptográficas em hardware.
- **Exemplo:** Gerenciamento de chaves criptográficas para transações financeiras em uma bolsa de valores.

### AWS Shield
O AWS Shield protege aplicativos contra ataques DDoS.
- **Exemplo:** Proteção de um site de notícias contra ataques DDoS durante eventos de alta visibilidade.

### AWS WAF
O AWS WAF protege aplicativos web contra explorações comuns da web.
- **Exemplo:** Proteção de um portal de e-commerce contra ataques de injeção SQL.

### AWS Macie
O AWS Macie usa machine learning para identificar e proteger dados sensíveis.
- **Exemplo:** Monitoramento de dados confidenciais armazenados no S3 para evitar vazamentos de informações.

### AWS GuardDuty
O AWS GuardDuty detecta atividades maliciosas e anomalias para proteger contas e workloads da AWS.
- **Exemplo:** Detecção de comportamentos suspeitos em instâncias EC2 de um banco online.

### AWS Inspector
O AWS Inspector avalia a segurança de aplicativos, procurando vulnerabilidades ou desvios das melhores práticas.
- **Exemplo:** Avaliação de segurança de uma aplicação web antes do lançamento para garantir conformidade com padrões de segurança.

### AWS Secrets Manager
O AWS Secrets Manager gerencia e recupera credenciais de acesso e outros segredos.
- **Exemplo:** Armazenamento seguro de credenciais de banco de dados usadas por uma aplicação de gerenciamento financeiro.

### AWS Config
O AWS Config permite avaliar, auditar e avaliar as configurações de recursos da AWS.
- **Exemplo:** Monitoramento de conformidade das configurações de segurança de recursos AWS em uma empresa de saúde.

### AWS Security Hub
O AWS Security Hub oferece uma visão abrangente das posturas de segurança em toda a conta AWS.
- **Exemplo:** Centralização e correlação de alertas de segurança em uma grande empresa de tecnologia.

### AWS Artifact
O AWS Artifact oferece acesso a relatórios de conformidade da AWS e outros documentos relacionados.
- **Exemplo:** Obtenção de relatórios de conformidade para auditorias de segurança e conformidade regulatória.

### AWS Audit Manager
O AWS Audit Manager automatiza a coleta de evidências para auditorias de conformidade.
- **Exemplo:** Preparação para auditorias de conformidade com regulamentações como GDPR e HIPAA.

### AWS Single Sign-On (SSO)
O AWS Single Sign-On (SSO) permite gerenciar o acesso centralizado a várias contas e aplicativos da AWS.
- **Exemplo:** Fornecimento de um único ponto de acesso para usuários de uma organização com várias contas AWS.

- **Exemplo:** Auditoria de atividades de usuários e serviços para garantir a conformidade em uma organização financeira.

### AWS Systems Manager
O AWS Systems Manager oferece um hub de gerenciamento para recursos na AWS e em ambientes híbridos.
- **Exemplo:** Automação de atualizações de software e patches em instâncias EC2 para uma empresa global de tecnologia.

### AWS License Manager
O AWS License Manager facilita o gerenciamento de licenças de software.
- **Exemplo:** Rastreamento e otimização do uso de licenças de software em uma grande organização.

### AWS Managed Services
O AWS Managed Services oferece recursos proativos, preventivos e de detecção para operações na nuvem.
- **Exemplo:** Gerenciamento de operações de nuvem para uma empresa que deseja focar na inovação em vez da infraestrutura.

### AWS Budgets
O AWS Budgets permite definir orçamentos personalizados para monitorar e gerenciar custos e uso na AWS.
- **Exemplo:** Definição de alertas de orçamento para evitar gastos excessivos em uma conta AWS.

### AWS Cost Explorer
O AWS Cost Explorer permite visualizar e analisar custos e uso da AWS ao longo do tempo.
- **Exemplo:** Análise de tendências de custos mensais para otimização de gastos em uma empresa de mídia.

## Integração de Aplicações

### Amazon SNS
O Amazon SNS é um serviço de mensagem pub/sub que facilita a comunicação entre aplicativos distribuídos.
- **Exemplo:** Envio de notificações de eventos para usuários de um aplicativo móvel.

### Amazon SQS
O Amazon SQS é um serviço de fila de mensagens que desacopla e dimensiona microsserviços.
- **Exemplo:** Filas de tarefas para processamento assíncrono em um sistema de processamento de imagens.

### Amazon EventBridge
O Amazon EventBridge facilita a criação de arquiteturas orientadas a eventos em escala.
- **Exemplo:** Roteamento de eventos entre diferentes microserviços em uma plataforma de e-commerce.

### Amazon API Gateway
O Amazon API Gateway permite a criação, publicação, manutenção, monitoração e proteção de APIs em grande escala.
- **Exemplo:** API RESTful para um aplicativo móvel que se conecta a serviços back-end.

### AWS Step Functions
O AWS Step Functions orquestra serviços da AWS, automatizando processos de negócios com fluxos de trabalho visuais.
- **Exemplo:** Orquestração de processos de ETL para movimentação e transformação de dados.

### AWS AppSync
O AWS AppSync facilita a criação de APIs GraphQL para sincronização de dados em tempo real.
- **Exemplo:** API para um aplicativo de chat que sincroniza mensagens entre usuários em tempo real.

### AWS CodeCommit
O AWS CodeCommit é um serviço de controle de versão que hospeda repositórios Git seguros e escaláveis.
- **Exemplo:** Repositório de código para desenvolvimento colaborativo de um novo aplicativo.

### AWS CodePipeline
O AWS CodePipeline automatiza pipelines de entrega contínua para atualizações rápidas e confiáveis de aplicativos e infraestruturas.
- **Exemplo:** Pipeline CI/CD para automatizar testes e implantações de um aplicativo web.

### AWS CodeBuild
O AWS CodeBuild compila código-fonte, executa testes e produz pacotes de software prontos para implantação.
- **Exemplo:** Compilação e teste automatizado de código para um aplicativo de análise de dados.

### AWS CodeDeploy
O AWS CodeDeploy automatiza a implantação de código em qualquer instância, incluindo Amazon EC2 e servidores on-premises.
- **Exemplo:** Implantação automatizada de atualizações de software para servidores de produção.

### AWS CodeStar
O AWS CodeStar gerencia o ciclo de vida de desenvolvimento de aplicativos, integrando-se com ferramentas populares de desenvolvimento.
- **Exemplo:** Plataforma integrada para gerenciar o desenvolvimento de um novo serviço de assinatura.

## DevOps

### AWS OpsWorks
O AWS OpsWorks é um serviço de automação de configuração que usa Chef e Puppet para gerenciar infraestruturas em nuvem e on-premises.
- **Exemplo:** Automação de configuração e gerenciamento de servidores web em um ambiente de produção.

### AWS CloudFormation
O AWS CloudFormation permite modelar e configurar seus recursos da AWS para que você possa passar menos tempo gerenciando esses recursos e mais tempo se concentrando nos aplicativos executados neles.
- **Exemplo:** Criação de uma pilha de recursos para um ambiente de desenvolvimento, incluindo VPC, instâncias EC2 e RDS.

### AWS Service Catalog
O AWS Service Catalog permite que as organizações criem e gerenciem catálogos de serviços de TI aprovados para uso na AWS.
- **Exemplo:** Catálogo de serviços de TI para desenvolvedores de uma empresa de software utilizarem para criar novos ambientes de desenvolvimento.

### AWS Professional Services
O AWS Professional Services oferece consultoria especializada para ajudar as empresas a migrar e operar na AWS.
- **Exemplo:** Assistência na migração de um data center local para a nuvem AWS com as melhores práticas.

### AWS Support
O AWS Support oferece suporte técnico e orientação em relação ao uso dos serviços da AWS.
- **Exemplo:** Suporte técnico para resolução de problemas e otimização de serviços AWS.

### AWS Training and Certification
O AWS Training and Certification oferece treinamento e certificação para ajudar indivíduos e equipes a desenvolver suas habilidades na AWS.
- **Exemplo:** Programas de treinamento para equipes de TI em uma empresa que está adotando a AWS.

## Ferramentas de Migração e Transferência

### AWS Migration Hub
O AWS Migration Hub fornece uma localização central para monitorar o progresso das migrações de aplicações para a AWS.
- **Exemplo:** Monitoramento da migração de vários aplicativos empresariais para a AWS.

### AWS Database Migration Service (DMS)
O AWS Database Migration Service ajuda a migrar bancos de dados para a AWS de forma rápida e segura.
- **Exemplo:** Migração de um banco de dados Oracle on-premises para o Amazon RDS.

### AWS Server Migration Service (SMS)
O AWS Server Migration Service facilita a migração de cargas de trabalho de servidores on-premises para a AWS.
- **Exemplo:** Migração de servidores VMware para instâncias do Amazon EC2.
