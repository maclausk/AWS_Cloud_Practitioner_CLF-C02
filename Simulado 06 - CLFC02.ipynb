**Pergunta 1**  
Qual serviço permite executar cargas de trabalho HPC (High Performance Computing) de maneira eficiente?  
a) Batch  
b) **EC2**  
c) Lambda  
d) Elastic Beanstalk

**Explicação:**  
**Correta:**  
- O Amazon EC2 fornece capacidade computacional escalável na nuvem e permite executar cargas de trabalho HPC de maneira eficiente, fornecendo instâncias otimizadas para HPC com recursos específicos para acelerar o desempenho.

- O AWS Batch é um serviço que permite executar em lote cargas de trabalho em larga escala na nuvem, mas não é específico para cargas de trabalho HPC.

- O Elastic Beanstalk é um serviço que facilita a implantação e o gerenciamento de aplicativos web, mas não é otimizado para cargas de trabalho HPC.

**Pergunta 4**  
Os componentes das soluções do AWS Virtual Private Network (VPN) permitem conexões seguras entre redes locais, escritórios remotos, dispositivos de clientes e a rede global da AWS. Quais são os serviços que compõem essas soluções? (selecione 2 alternativas)  
a) AWS Connect  
b) AWS Direct Connect  
c) **AWS Client VPN**  
d) **AWS Site-to-Site VPN**  
e) AWS Storage Gateway

**Explicação:**  
**Correta:**  
- O AWS VPN é constituído por dois serviços principais: o AWS Site-to-Site VPN e o AWS Client VPN. O Site-to-Site VPN estabelece túneis criptografados entre sua rede e as Amazon Virtual Private Clouds ou os AWS Transit Gateways. O AWS Client VPN conecta usuários à AWS ou a recursos locais por meio de um cliente de software VPN.

**Incorretas:**  
- O Direct Connect cria uma conexão dedicada que não trafega dados fora da infraestrutura da AWS.  
- O Connect permite a implementação de uma central de atendimento multi-canal.  
- O Storage Gateway é um serviço de armazenamento em nuvem híbrida.
