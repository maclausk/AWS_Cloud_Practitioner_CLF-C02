**Pergunta 1:** O que pode-se afirmar sobre o AWS Budget? (Selecione 3 alternativas)  
a) **Podem ser utilizados para criar e gerenciar orçamentos**  
b) **Podem ser utilizados para enviar alertas para te ajudar no controle do uso dos serviços da sua conta**  
c) **Podem ser utilizados para refinar e filtrar orçamentos**  
d) Podem enviar notificações para até 5 destinatários por email  

**Explicação:**  
**Corretas:**  
- **Podem ser utilizados para criar e gerenciar orçamentos:** O AWS Budgets permite que os usuários criem orçamentos personalizados com base nos custos e uso. Isso ajuda a prever e gerenciar despesas com eficiência.  
- **Podem ser utilizados para enviar alertas para te ajudar no controle do uso dos serviços da sua conta:** Os usuários podem configurar alertas para serem notificados quando os custos ou o uso atingirem limites pré-definidos, permitindo uma gestão proativa das despesas.  
- **Podem ser utilizados para refinar e filtrar orçamentos:** O serviço oferece opções para filtrar e ajustar os orçamentos, proporcionando uma visão detalhada dos gastos em diferentes serviços ou contas.  
**Incorreta:**  
- **Podem enviar notificações para até 5 destinatários por email:** O AWS Budgets permite o envio de notificações para até 10 destinatários, não apenas 5, garantindo uma ampla comunicação dentro da equipe.

**Pergunta 2:** Uma empresa está desenvolvendo uma nova plataforma para a distribuição global de vídeos. Qual serviço pode reduzir a latência entre a solicitação e a entrega do serviço para essa solução?  
a) Amazon S3  
b) Amazon ElasticCache  
c) Amazon Redis  
d) **Amazon CloudFront**

**Explicação:**  
**Correta:**  
- **Amazon CloudFront:** É uma rede de entrega de conteúdo (CDN) que distribui conteúdo globalmente usando uma rede de edge locations. Isso reduz a latência, pois os dados são entregues de um ponto mais próximo ao usuário final, melhorando a velocidade e a experiência do usuário.  
**Incorretas:**  
- **Amazon S3:** É um serviço de armazenamento de objetos que, embora eficiente para armazenamento, não otimiza a latência da entrega de conteúdo como o CloudFront.  
- **ElasticCache e Redis:** São soluções para cache de dados de aplicativos, como sessões de usuário ou consultas frequentes a bancos de dados, e não são ideais para entrega de conteúdo multimídia.

**Pergunta 4:** Clientes de uma empresa na Europa e na Austrália expressaram insatisfação devido à demora no carregamento de imagens em um site, resultando em uma experiência ruim. Os recursos de infraestrutura desse site estão provisionados na Região de São Paulo. O que poderia ser feito imediatamente para aprimorar essa situação?
a) Habilitar o Intelligent-Tiering das classes do S3  
b) Criar uma conta nessas regiões e duplicar a infraestrutura  
c) Deixar as imagens numa resolução bem baixa  
d) **Habilitar o CloudFront**

**Explicação:**  
**Correta:**  
- **Habilitar o CloudFront:** Este serviço faz o cache do conteúdo em edge locations, que são mais próximas dos usuários finais. Isso melhora significativamente o tempo de resposta e reduz a latência, melhorando a experiência do usuário.  
**Incorretas:**  
- **Intelligent-Tiering do S3:** É usado para otimização de custos de armazenamento com base na frequência de acesso, mas não aborda problemas de latência.  
- **Criar contas nessas regiões:** Embora possa melhorar o acesso, é uma solução complexa e cara, e nem sempre necessária com o uso adequado do CloudFront.  
- **Reduzir a resolução das imagens:** Isso pode comprometer a qualidade visual e não é uma solução ideal para melhorar a velocidade de carregamento.

**Pergunta 5:** Qual serviço da AWS é empregado para realizar migrações rápidas e seguras de dados entre bancos de dados?  
a) AWS Snow  
b) AWS Glue  
c) AWS Connect  
d) **AWS Data Migration Service (DMS)**

**Explicação:**  
**Correta:**  
- **AWS Data Migration Service (DMS):** Este serviço facilita a migração de bancos de dados para a AWS, com interrupção mínima do serviço. Ele suporta migrações homogêneas (como Oracle para Oracle) e heterogêneas (como Oracle para PostgreSQL), permitindo que dados sejam transferidos com segurança e eficiência.  
**Incorretas:**  
- **AWS Snow:** É usado para transferências físicas de grandes volumes de dados entre locais.  
- **AWS Glue:** É um serviço de integração de dados para processos de ETL (extrair, transformar e carregar).  
- **AWS Connect:** É um serviço de central de contato para atendimento ao cliente, não para migração de dados.

**Pergunta 6:** Uma startup precisa assegurar a alta disponibilidade de uma aplicação que será lançada no Brasil. Qual seria a estratégia mais eficaz para utilizar o AWS RDS atendendo a esse requisito?
a) Multi Regions  
b) Criar um Security Group para isolar apenas os bancos de dados  
c) Read Replicas  
d) **Multi A-Z (Distribuir os bancos em mais de uma zona de disponibilidade)**

**Explicação:**  
**Correta:**  
- **Multi A-Z (Availability Zones):** Distribuir instâncias de banco de dados em múltiplas zonas de disponibilidade dentro da mesma região garante que, se uma zona falhar, as outras possam continuar a operar, oferecendo alta disponibilidade e proteção contra falhas em uma única zona.  
**Incorretas:**  
- **Multi Region:** É mais adequado para recuperação de desastres (DR) e pode aumentar a latência, sendo menos eficiente para alta disponibilidade.  
- **Read Replicas:** Melhoram a escalabilidade de leitura, mas não garantem alta disponibilidade.  
- **Security Groups:** Controlam o acesso de rede, mas não impactam diretamente na disponibilidade do banco de dados.

b) Não possui suporte a conteinerização

**Pergunta 8:** Por que é necessário associar um Elastic Block Store (EBS) a uma instância EC2?  
a) Para deixar o fluxo de transmissão de dados mais rápido.  
b) **Para fornecer armazenamento persistente e escalável.**  
c) Para permitir redundância computacional.  
d) Para que os dados sejam criptografados.

**Explicação:**  
**Correta:**  
- **Para fornecer armazenamento persistente e escalável:** O EBS oferece armazenamento persistente que não é perdido quando a instância EC2 é interrompida, além de ser escalável e permitir a configuração de volumes adicionais conforme necessário.  
**Incorretas:**  
- **Para deixar o fluxo de transmissão de dados mais rápido:** O desempenho depende do tipo de volume EBS escolhido, mas sua principal função é armazenamento.  
- **Para permitir redundância computacional:** Isso é alcançado com outros mecanismos, como Load Balancing ou Auto Scaling.  
- **Para que os dados sejam criptografados:** Embora o EBS ofereça criptografia, a segurança dos dados é uma das suas funções, mas não a principal razão para usá-lo com EC2.
